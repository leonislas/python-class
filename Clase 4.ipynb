```python
def segundo_orden(y,t):
    return array([y[1], -y[0]]) # define la ecuacion diferencial
```

```python
def euler(func, p_ini, t_final, h):
    pasos = arange(p_ini[0],t_final+h,h) #makes the domain of the solution
    y = [0]*len(pasos)
    y[0] = p_ini[1] # va a poner un vector de dos elementos en y[0] por la operacion anterior
    for i in range(len(pasos)-1):
        y[i+1] = y[i] + func(y[i],pasos[i])*h
    sol = pasos, y
    return sol
```

```python
sol = euler(segundo_orden,[0,[0,1]], 20, 0.02)
y = zeros(len(sol[0]))
#print sol[1]
for i in range(len(sol[1])):
    y[i] = sol[1][i][1]
          
plot(sol[0], y)
plot(sol[0], cos(sol[0]))
```
