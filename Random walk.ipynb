```julia
function trayectoria(N=1000)
    x = y = 0.0
    data = zeros(N, 2)
    
    for i in 1:N
        x += randn()
        y += randn()

        data[i, :] = [x, y]
    end
    
    return data
end
    
```

```julia
trayecs = [trayectoria(100) for i in 1:10000];
```

```julia
p = plot()

for i in 1:length(trayecs)
    plot!(trayecs[i][:,1], trayecs[i][:,2])
end
```
